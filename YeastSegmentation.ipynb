```python
from matplotlib import image
from matplotlib import pyplot
dir = '/content/drive/My Drive/'
image1 = image.imread('/content/drive/My Drive/Yeast/01.JPG')
image2 = image.imread('/content/drive/My Drive/Yeast/02.JPG')
image3 = image.imread('/content/drive/My Drive/Yeast/03.JPG')
image4 = image.imread('/content/drive/My Drive/Yeast/04.JPG')
```

```python
pyplot.imshow(image1)
pyplot.show()
pyplot.imshow(image2)
pyplot.show()
pyplot.imshow(image3)
pyplot.show()
pyplot.imshow(image4)
pyplot.show()
```

## IMAGE 1

```python
plt.figure(figsize=(9,9))
plt.title('original', fontsize=18)
plt.imshow(image1)
plt.show()

grayscale = rgb2gray(image1)
plt.figure(figsize=(9,9))
plt.title('edge segmentation', fontsize=18)
edges = skimage.feature.canny(grayscale, sigma=3)
plt.imshow(edges)

fig, ax = plt.subplots(figsize=(9,9))
ax.set_title('Contour Segementation', fontsize=18)
qcs = ax.contour(grayscale, origin='image')
plt.show()
```
